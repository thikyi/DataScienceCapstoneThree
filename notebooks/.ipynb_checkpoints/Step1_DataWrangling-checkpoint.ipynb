#### Handling Missing Data

```python
#Number Of Missing Values By Column
missingdata= pd.concat([rawdata.isnull().sum(), 100 * rawdata.isnull().mean()], axis=1)
missingdata.columns=['count', '%']
missingdata.sort_values(by=['count'])
```

The most missing entries are AgeHH1, AgeHH2, PercChangeMinutes and PercChangeRevenues and account for less than 2% of the data.

```python
# Fill missing values

rawdata['AgeHH1'] = rawdata['AgeHH1'].fillna(rawdata['AgeHH1'].median())
rawdata['AgeHH2'] = rawdata['AgeHH2'].fillna(rawdata['AgeHH2'].median())
rawdata['PercChangeMinutes'] = rawdata['PercChangeMinutes'].fillna(rawdata['PercChangeMinutes'].median())
rawdata['PercChangeRevenues'] = rawdata['PercChangeRevenues'].fillna(rawdata['PercChangeRevenues'].median())
rawdata['MonthlyRevenue'] = rawdata['MonthlyRevenue'].fillna(rawdata['MonthlyRevenue'].median())
rawdata['MonthlyMinutes'] = rawdata['MonthlyMinutes'].fillna(rawdata['MonthlyMinutes'].median())
rawdata['TotalRecurringCharge'] = rawdata['TotalRecurringCharge'].fillna(rawdata['TotalRecurringCharge'].median())
rawdata['DirectorAssistedCalls'] = rawdata['DirectorAssistedCalls'].fillna(rawdata['DirectorAssistedCalls'].median())
rawdata['OverageMinutes'] = rawdata['OverageMinutes'].fillna(rawdata['OverageMinutes'].median())
rawdata['RoamingCalls'] = rawdata['RoamingCalls'].fillna(rawdata['RoamingCalls'].median())
rawdata['ServiceArea'] = rawdata['ServiceArea'].fillna(rawdata['ServiceArea'].mode()[0])
rawdata['Handsets'] = rawdata['Handsets'].fillna(rawdata['Handsets'].median())
rawdata['HandsetModels'] = rawdata['HandsetModels'].fillna(rawdata['HandsetModels'].median())
rawdata['CurrentEquipmentDays'] = rawdata['CurrentEquipmentDays'].fillna(rawdata['CurrentEquipmentDays'].median())
```

#### Check if there is any duplicate records

```python
duplicate_records = rawdata[rawdata.duplicated()]

if duplicate_records.empty:
    print("No duplicate records found.")
else:
    print("Duplicate records found:")
    print(duplicate_records)
```

```python
# List of columns to visualize for outliers
columns_to_visualize = ['MonthlyRevenue', 'MonthlyMinutes', 'TotalRecurringCharge', 'DirectorAssistedCalls', 'OverageMinutes']

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Create box plots for each column
for i, column in enumerate(columns_to_visualize, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=rawdata[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

There are a few ways to handle outliers and use a statistical method to identify them, such as:</br>

Z-Scores: Data points with a z-score greater than 3 or less than -3 are considered outliers.</br>
Interquartile Range (IQR): Outliers are identified as observations that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR, where Q1 and Q3 are the first and third quartiles, respectively.</br>
Once identified, you can handle outliers in several ways:</br>
Remove: Discard outliers from your dataset.</br>
Cap: Set outliers to a specific maximum or minimum value.</br>
Impute: Replace outliers with a statistically derived value, such as the mean or median of the non-outlier values.</br>
Adjust: Use a transformation to reduce the impact of outliers.</br>

<b>The presence of outliers can be seen in all selected variables, especially in "MonthlyMinutes" and "OverageMinutes," which show a considerable number of points lying far above the upper quartile, indicating unusually high values compared to the rest of the data.</b> </br>
<b>Given the variability in telecom data resulting from diverse customer lifestyles, it is essential to recognize that outliers may vary from one customer to another. Consequently, the decision to remove or impute outliers should be approached with caution and may necessitate careful consideration. With that, no further change required to data.</b>
