#### Hypothesis Testing

To further investigate the statistical significance of the findings, let's define and conduct a null hypothesis test. A common hypothesis related to churn could be:<
</br>
Null Hypothesis : There is no significant difference in the mean value between churned and non-churned customers.</br>
Alternative Hypothesis: There is a significant difference in the mean value between churned and non-churned customers.</br>

##### Behavioral Influence on Churn

```python
from scipy import stats  # Import the stats module from scipy

beh_features_to_test = [
    'MonthlyRevenue',
    'MonthlyMinutes',
    'TotalRecurringCharge',
    'DirectorAssistedCalls',
    'OverageMinutes',
    'RoamingCalls',
    'CustomerCareCalls'
]

# Perform t-tests for each feature
hypothesis_test_results = {}

for feature in beh_features_to_test:
    churned_group = df[df['Churn'] == 1][feature]
    non_churned_group = df[df['Churn'] == 0][feature]
    t_stat, p_value = stats.ttest_ind(churned_group, non_churned_group, equal_var=False)
    hypothesis_test_results[feature] = {'t_stat': t_stat, 'p_value': p_value}

# Prepare the summary of findings as a list
findings_summary = [
    f"{feature}: Mean difference is {'significant' if result['p_value'] < 0.05 else 'not significant'} (t-statistic = {result['t_stat']:.3f}, p-value = {result['p_value']:.5f})"
    for feature, result in hypothesis_test_results.items()
]

findings_summary
```

##### Demographical Influence on Churn

```python
from scipy.stats import chi2_contingency


# List of categorical features to perform chi-square tests on
categorical_features = ['IncomeGroup', 'MaritalStatus', 'Occupation', 'CityCode','ChildrenInHH']

# Initialize a dictionary to store test results
chi_square_results = {}

# Perform chi-square tests
for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df['Churn'])
    
    # Perform the chi-square test
    chi2, p_value, dof, _ = chi2_contingency(contingency_table)
    
    # Store the results
    chi_square_results[feature] = {
        'Chi-square Statistic': chi2,
        'p-value': p_value,
        'Degrees of Freedom': dof
    }

# Output the results
for feature, result in chi_square_results.items():
    print(f"{feature}:")
    print(f"Chi-square Statistic: {result['Chi-square Statistic']}, p-value: {result['p-value']}")
    print(f"{feature} Mean difference is {'significant' if result['p-value'] < 0.05 else 'not significant'}")
```
